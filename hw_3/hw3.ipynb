# **Выведите самое популярное мужское и самое популярное женское имя на корабле**

```python
def parse_first_name(name):
    first = name.str.extract(r"Mrs\.\s+[^(]*\((\w+)", expand=False)
    first.loc[first.isna()] = name.str.extract(r"\.\s+(\w+)", expand=False)
    return first
```

```python
female_names = parse_first_name(df.loc[df["Sex"]=="female","Name"]).value_counts()
```

```python
female_names[0:1]
```

```python
male_names = parse_first_name(df.loc[df["Sex"]=="male","Name"]).value_counts()
male_names[0:1]
```

# **Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе**

```python
i=0
for i in df['Pclass'].unique():
  df1=df.loc[df["Pclass"]==i]
  male_names = parse_first_name(df1.loc[df1["Sex"]=="male","Name"]).value_counts()

  print("в  классе",i,"самое популярное мужское",male_names[0:1])
```

```python
i=0
for i in df['Pclass'].unique():
  df1=df.loc[df["Pclass"]==i]
  female_names = parse_first_name(df1.loc[df1["Sex"]=="female","Name"]).value_counts()

  print("в  классе",i,"самое популярное мужское",female_names[0:1])
```
